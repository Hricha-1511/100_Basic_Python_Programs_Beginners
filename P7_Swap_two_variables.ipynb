# Method 1
#Swapping Using Third Variable aka temporary variable

```python
x = 13
y = 12

temp = x
print("The value of the temp variable is", temp)

x= y
print("The value of x is:", x)

y=temp
print("The value of y is :", y)
```

# Method 2
#Swapping without using temp variable

```python
x = 12
y = 13

x, y = y, x

print("The value of x is:", x)
print("The value of y is:", y)
```
