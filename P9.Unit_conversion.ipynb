```python
km = float(input("Enter the value in Km:"))

miles =  (0.621371)* km

print(km, "kms in miles will be ", miles)
```

```python
# Celcius to fahrenheit

# 0 deg is equal to 32 Fahrenhiet

# 40 deg Celcius to Fahrenhiet

# (C+9/5)+ 32 = F
```

```python
celcius = int(input("Enter the value in celcius:"))


fahrenhiet = (celcius*(9/5))+32

print("The converted value is:", fahrenhiet,"Fahrenhiet")
```
