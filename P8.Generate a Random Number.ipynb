```python
import random
```

```python
# Inclusive number between  limits
num = random.randint(0,10)
print("A random number between 0 and 10 is:", num)
```

```python
  # Random float:  0.0 <= x < 1.0
random.random()
```

```python
  # Random float:  2.5 <= x <= 10.0
random.uniform(2.5, 10.0)
```

```python
# Interval between arrivals averaging 5 seconds
random.expovariate(1 / 5)
```

```python
 # Integer from 0 to 9 inclusive
random.randrange(10)
```

```python

  # Even integer from 0 to 100 inclusive
random.randrange(0, 101, 2)
```

```python
#  This gives a random integer from 1 to 9 (10 is excluded).
num = random.randrange(1, 10)
print(num)
```

```python

# Single random element from a sequencerandom.
random.choice(['win', 'lose', 'draw'])
```

```python
# Shuffle a list



deck = 'ace two three four'.split()
random.shuffle(deck)
deck
['four', 'two', 'ace', 'three']
```

```python
# Four samples without replacement
random.sample([10, 20, 30, 40, 50], k=4)


```
